# Q2) A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


```python
import pandas as pd
```

```python
df=pd.read_csv('LabTAT.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
l1=df['Laboratory 1']
```

```python
l2=df['Laboratory 2']
```

```python
l2.hist()
```

# Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('BuyerRatio.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df.plot(kind='bar')
```

```python
df=df.iloc[:,1:6]
df
```

```python
val=chi2_contingency(df)
```

```python
pvalue=val[1]
pvalue
```

```python
if pvalue <= 0.05:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```
